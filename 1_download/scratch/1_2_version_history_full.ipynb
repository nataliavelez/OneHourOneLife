The version history in `1_2` was incomplete, so now we're using the [changeLog](https://github.com/jasonrohrer/OneLife/blob/master/documentation/changeLog.txt) in the [OneHourOneLife Github repository](https://github.com/jasonrohrer/OneLife) as a source.

```python
import re,requests
import pandas as pd
```

Download text file:

```python
log_url = 'https://raw.githubusercontent.com/jasonrohrer/OneLife/master/documentation/changeLog.txt'
log_request = requests.get(log_url)
log_txt = log_request.text

print(log_txt)
```

Extract version numbers and dates from log:

```python
search_txt = 'Version ([0-9]+)\s+([0-9]+-[A-Za-z]+-[0-9]+)'
log_search = re.findall(search_txt, log_txt)
```

```python
log_search
```

Save search results as a dataframe:

```python
log_df = pd.DataFrame(log_search, columns=['version', 'release_date'])
log_df.version = pd.to_numeric(log_df.version)
log_df.release_date = pd.to_datetime(log_df.release_date)
log_df.head()
```

Save to file:

```python
log_df.to_csv('outputs/version_history_full.tsv', sep='\t', index=None)
```
